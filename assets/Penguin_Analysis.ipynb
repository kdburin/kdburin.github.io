## Abstract
This report determines which characteristics in the dataset are effective for distinguishing between species of penguins, like Adelie, Chinstrap, and Gentoo. Rather than focusing only on prediction accuracy, I study how much information each variable carries about species identity. For each candidate variable, I examine its distribution, define bins using hierarchical clustering, construct species-by-bin contingency tables, and compute entropy-based separation measures. This allows me to compare variables by how much uncertainty about species they remove and to identify which measurements are most biologically informative.

## Method 
* Select one or more variables for analysis
* Plot the ECDF of the selected variable to examine its distribution and identify potential clusters in the data.
* Apply hierarchical clustering in order to identify natural clusters. 
* Use the clusters from the dendrogram as bins.
* Build a species × bin contingency table. 
Rows = penguin species
Columns = bins
Entries = how many penguins of each species fall into each bin
* Divide each row entry by its corresponding row total in the contingency table to get the probability vector.
Such probability vectors indicate how each species is allocated to bins and act as the observed (or alternative) multinomial probability vectors.
* Compute the pooled null vector p0 by adding the column totals across all species and dividing by the total sample size.
This represents how all penguins fall across the bins if species labels are ignored.
* Two multinomial distributions can be defined for each species:
Observed/alternative: Probability vector corresponding to the species row Null: Pooled null vector p0. The number of observations must equal the size of the species row.
* Simulate many multinomial samples under both the observed and null models, keeping the sample size fixed at that species’ row total.
* For each simulated sample, convert the count vector to proportions (a probability vector), then compute entropy.
* Plot and compare the entropy distribution of the observation with the entropy distribution of the null hypothesis. The more overlap there is, the weaker the feature, and the less overlap, the stronger the discrimination between the species.

### Separation Score

The idea behind the separation score is to measure how much uncertainty about species is removed after knowing a variable. In this report, I use the separation score as one summary measure of how informative a variable is for distinguishing species. A larger separation score means that the variable reduces more species uncertainty and therefore gives better separation.

Although having the same number of clusters for each variable would make the separation scores easier to compare, in this analysis each variable was split according to its own natural tendency to form clusters. Some variables naturally split into more clusters than others. However, this does not make the separation score incorrect. Rather, it means that the score is based on the amount of species separation that can be identified from each variable’s natural clustering structure.

$$
S(A) = 1 - \frac{H(Y \mid A)}{H(Y)}
$$

where:

- $Y$ = species label  
- $A$ = binned variable  
- $H(Y)$ = entropy of species  
- $H(Y \mid A)$ = conditional entropy of species given bins $A$


* Apply the same technique to all potential predictor variables and to selected interaction/combination predictor variables.
* Compare the separation scores obtained by different predictors.
* 
* Select the most valuable predictors and form the prediction model using them.


## ECDF and Histograms 


```python

x = df[["bill_length_mm"]]

Z = linkage(x, method="complete", metric = "euclidean")  

plt.figure(figsize=(9, 4))
dendrogram(Z, no_labels=True)
plt.title("Dendrogram: Bill Length")
plt.xlabel("Observations")
plt.ylabel("Distance")
plt.show()
label = fcluster(Z, t=7, criterion="distance")
df2 = df[["bill_length_mm", "species"]].copy()
df2["bill_length_mm_cluster"] = label
h1 = pd.crosstab(df2["species"], df2["bill_length_mm_cluster"])


plt.figure()
sns.ecdfplot(x = "bill_length_mm", data = df)
plt.xlabel("Bill Length in mm")
plt.title("Empirical CDF")
plt.show()


def cluster_bin_hist(dfin, value_col, cluster_col):
    d = dfin[[value_col, cluster_col]].dropna().copy()
    g = (d.groupby(cluster_col)[value_col].agg(center="median", count="size").sort_values("center"))
    centers = g["center"].to_numpy()
    mids = (centers[:-1] + centers[1:]) / 2
    left = np.r_[d[value_col].min(), mids]
    right = np.r_[mids, d[value_col].max()]
    g["left"] = left
    g["right"] = right
    g["width"] = g["right"] - g["left"]
    n = g["count"].sum()
    g["density"] = (g["count"] / n) / g["width"]  
    plt.figure(figsize=(8,5))
    plt.bar(g["left"], g["density"], width=g["width"], align="edge", edgecolor="black", color = "LightBlue")
    for cid, r in g.iterrows():
        x = (r["left"] + r["right"]) / 2
        y = r["density"] * 0.08
        plt.text(x, y, str(cid), ha="center", va="bottom", fontsize=12)
    plt.xlabel(value_col)
    plt.ylabel("density")
    plt.title(f"{value_col}: Cluster-Bin Histogram")
    plt.show()
    return g  

cluster_bin_hist(df2, "bill_length_mm", "bill_length_mm_cluster")
h1
```

## Bill Length analysis

```python
np.random.seed(1)

def sim_entropy(n, probs, n_sim=10000):
    z = np.random.multinomial(n, probs, size=n_sim)
    ent = []
    for row in z:
        p = row / row.sum()
        h = -np.sum(p[p > 0] * np.log(p[p > 0]))   
        ent.append(h)
    return np.array(ent)


vector_p0 = (h1.sum(axis=0) / h1.values.sum()).to_list()


for species in h1.index:
    row = h1.loc[species, :]
    n_species = int(row.sum())
    p_species = (row / n_species).to_list()

    ent_null = sim_entropy(n_species, vector_p0)
    ent_species = sim_entropy(n_species, p_species)

    plt.figure(figsize=(8,5))
    plt.hist(ent_null, bins=40, edgecolor="black", alpha=0.8, color="red")
    plt.hist(ent_species, bins=40, edgecolor="white", alpha=0.8, color="blue")
    plt.xlabel("Entropy")
    plt.ylabel("Count")
    plt.title(f"Distribution of Entropy Across Simulations {species}")
    plt.show()
h1
```

From the graphs above, we can see that the observed entropy distributions for all three species are clearly separated from the null entropy distributions. This suggests that bill length carries useful information for distinguishing species. Across the 10,000 simulated observations, the observed entropies are much lower than the null entropies, which means the species are distributed across the bill-length bins in a more structured way than we would expect by chance. Therefore, bill length appears to be a useful variable for separating penguin species.

## Island Analysis

```python
np.random.seed(1)

# 1) Contingency table for island
h_island = pd.crosstab(df["species"], df["island"])

p_null = (h_island.sum(axis=0) / h_island.values.sum()).to_numpy()

for sp in h_island.index:
    row = h_island.loc[sp].to_numpy()
    n_sp = int(row.sum())
    p_obs = row / n_sp

    ent_null = sim_entropy(n_sp, p_null)
    ent_obs = sim_entropy(n_sp, p_obs)

    mu_obs = ent_obs.mean()
    mu_null = ent_null.mean()
    sd_obs = ent_obs.std(ddof=1)
    sd_null = ent_null.std(ddof=1)

    gap = mu_null - mu_obs
    d = gap / np.sqrt((sd_obs**2 + sd_null**2) / 2)

    plt.figure(figsize=(8,5))
    plt.hist(ent_null, bins=40, edgecolor="black", alpha=0.8, color="red")
    plt.hist(ent_obs, bins=40, edgecolor="white", alpha=0.8, color="blue")
    plt.xlabel("Entropy")
    plt.ylabel("Count")
    plt.title(f"Entropy Simulation (island bins): {sp}")
    plt.show()
h_island
```

In the entropy graphs, it can be observed that the variable “island” behaves very differently depending on the penguin species. Since the entropy values for both Chinstrap and Gentoo are nearly zero, this indicates that each of these two species is found on only one island. On the other hand, Adelie is observed on all three islands, so its entropy is much higher and close to the maximum possible value for three groups. Thus, the variable “island” is highly discriminative for Chinstrap and Gentoo penguins, but less informative for Adelie.

# Analysis of Multiple Variables together: 

Motivation for choosing separation score is the fact that apart from describing the variables, it will also serve the purpose of knowing which variables have to be selected for predicting the species. Separation score is defined by the amount of uncertainty that is reduced regarding the species due to the variable or the combination of variables used. In simpler terms, it is the reduction of uncertainty in identifying the species based on the variable. The higher the separation score, the higher the amount of uncertainty that has been eliminated. Although there could be variations in the separability among species by different variables, the main purpose remains the same, which is to minimize the uncertainty among the species. Therefore, separation score gives an indication of the strength of a particular variable or combination of variables.

We use combinations of variables, compute the separation score for each one, and then choose the best combination for prediction. In this way, we can identify which combination removes the most entropy. 

```python
features = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
X_scaled = StandardScaler().fit_transform(df[features])

# Hierarchical clustering
Z = linkage(X_scaled, method="complete", metric="euclidean")

plt.figure(figsize=(10, 4))
dendrogram(Z, no_labels=True)
plt.title("Dendrogram: All Numerical Variables")
plt.xlabel("Observations")
plt.ylabel("Distance")
plt.show()


labels = fcluster(Z, t=2.7, criterion="distance")

# Separation table
h_all4 = pd.crosstab(df["species"], labels)
print("Contigency Table:")
h_all4
```

```python

p_null = (h_all4.sum(axis=0) / h_all4.values.sum()).to_numpy()

for sp in h_all4.index:
    row = h_all4.loc[sp].to_numpy()
    n_sp = int(row.sum())
    p_sp = row / n_sp

    ent_null = sim_entropy(n_sp, p_null)
    ent_sp = sim_entropy(n_sp, p_sp)

    plt.figure(figsize=(8,5))
    plt.hist(ent_null, bins=40, edgecolor="black", alpha=0.8, color="red")
    plt.hist(ent_sp, bins=40, edgecolor="white", alpha=0.8, color="blue")
    plt.xlabel("Entropy")
    plt.ylabel("Count")
    plt.title(f"Entropy Simulation: {sp}")
    plt.show()
```

The results indicate that there was a much higher degree of discrimination when all numerical variables were utilized at once compared to each individual variable. In the entropy graph, it can be seen that the entropy values for all three species are significantly below the null entropy distribution value, with no overlap whatsoever. Hence, it can be said that all numerical variables utilized give a pattern that is unlikely to happen randomly. As seen from the contingency table, the species known as Gentoo occupies the first two clusters only; Chinstrap, meanwhile, occupies most of the central clusters, while Adelie occupies only the last two clusters. ```python

```python
def H(counts):
    counts = np.asarray(counts, dtype=float)
    total = counts.sum()
    if total == 0:
        return 0.0
    p = counts[counts > 0] / total
    return -np.sum(p * np.log(p)) 

def score_table(h):
    n = h.to_numpy().sum()
    Hy = H(h.sum(axis=1).to_numpy())
    Hy_given_B = 0.0
    for col in h.columns:
        col_counts = h[col].to_numpy()
        nb = col_counts.sum()
        if nb > 0:
            Hy_given_B += (nb / n) * H(col_counts)

    separation_score = 1 - (Hy_given_B / Hy) if Hy > 0 else 0.0
    purity = h.max(axis=0).sum() / n
    return separation_score, purity

tables_named = {
    "bill_length": h1,
    "flipper_length": h2,
    "bill_depth": h3,
    "body_mass": h4,
    "all3_bill_length_bill_depth_flipper": h_all3,
    "all4_numeric": h_all4,
    "island": h_island,
    "all4_plus_island": h_all4_island
}

rows = []
for name, h in tables_named.items():
    sep, pur = score_table(h)
    rows.append({
        "model": name,
        "separation_score": sep,
        "purity": pur
    })

results_df = pd.DataFrame(rows).sort_values("separation_score", ascending=False).reset_index(drop=True)
results_df.round(4)


```

The separation score results provide additional support for the conclusions obtained from the clustering and entropy analyses. When the model using all four numerical variables was compared with the others, it showed the highest separation score and purity. For that reason, the four numerical variables were selected as the baseline model. Adding the variable island did not improve the separation score, even though it was still useful for prediction in a supervised model.

## Interacions
While the exact form of the interaction may be unknown, since there are many possible types of interactions between the variables, ranging from addition and multiplication to sine, cosine, and others, the most common convention for expressing the interaction between variables is by multiplying them. Therefore, multiplication was used here as an operational definition of the interaction. It is worth mentioning that the true form of interaction may not necessarily be multiplicative, but it is a plausible assumption.

```python

def make_h_table(X, title, criterion="maxclust", t=4):
    Z = linkage(X, method="complete", metric="euclidean")

    plt.figure(figsize=(10, 4))
    dendrogram(Z, no_labels=True)
    plt.title(f"Dendrogram: {title}")
    plt.xlabel("Observations")
    plt.ylabel("Distance")
    plt.show()

    labels = fcluster(Z, t=t, criterion=criterion)
    h = pd.crosstab(df["species"], labels)

    print(f"Contingency Table: {title}")
    print(h)
    return h

def plot_entropy_from_h(h, title):
    p_null = (h.sum(axis=0) / h.values.sum()).to_numpy()

    for sp in h.index:
        row = h.loc[sp].to_numpy()
        n_sp = int(row.sum())
        p_sp = row / n_sp

        ent_null = sim_entropy(n_sp, p_null)
        ent_sp = sim_entropy(n_sp, p_sp)

        plt.figure(figsize=(8, 5))
        plt.hist(ent_null, bins=40, edgecolor="black", alpha=0.8, color="red")
        plt.hist(ent_sp, bins=40, edgecolor="white", alpha=0.8, color="blue")
        plt.xlabel("Entropy")
        plt.ylabel("Count")
        plt.title(f"{title} - {sp}")
        plt.show()

X = (df["bill_length_mm"] * df["bill_depth_mm"]).to_numpy().reshape(-1, 1)
h_bl_bd = make_h_table(X, "bill_length_mm * bill_depth_mm", criterion="distance", t=200)
plot_entropy_from_h(h_bl_bd, "bill_length_mm * bill_depth_mm")

X = (df["bill_length_mm"] * df["flipper_length_mm"]).to_numpy().reshape(-1, 1)
h_bl_fl = make_h_table(X, "bill_length_mm * flipper_length_mm", criterion="distance", t=1500)
plot_entropy_from_h(h_bl_fl, "bill_length_mm * flipper_length_mm")

X = (df["bill_length_mm"] * df["body_mass_g"]).to_numpy().reshape(-1, 1)
h_bl_bm = make_h_table(X, "bill_length_mm * body_mass_g", criterion="maxclust", t=4)
plot_entropy_from_h(h_bl_bm, "bill_length_mm * body_mass_g")

X = (df["bill_depth_mm"] * df["flipper_length_mm"]).to_numpy().reshape(-1, 1)
h_bd_fl = make_h_table(X, "bill_depth_mm * flipper_length_mm", criterion="distance", t=400)
plot_entropy_from_h(h_bd_fl, "bill_depth_mm * flipper_length_mm")

X = (df["bill_depth_mm"] * df["body_mass_g"]).to_numpy().reshape(-1, 1)
h_bd_bm = make_h_table(X, "bill_depth_mm * body_mass_g", criterion="distance", t=10000)
plot_entropy_from_h(h_bd_bm, "bill_depth_mm * body_mass_g")

X = (df["flipper_length_mm"] * df["body_mass_g"]).to_numpy().reshape(-1, 1)
h_fl_bm = make_h_table(X, "flipper_length_mm * body_mass_g", criterion="distance", t=200000)
plot_entropy_from_h(h_fl_bm, "flipper_length_mm * body_mass_g")
```

From the interaction plots, the combinations bill length × body mass, bill length × flipper length, and flipper length × body mass all show useful species structure. In these cases, the observed entropy distributions are clearly lower than the null distributions, which suggests that these interactions are not behaving randomly. The contingency tables also show that species tend to concentrate in different clusters instead of being evenly mixed. So overall, these interactions seem to carry useful information for separating penguin species.

```python


interaction_pairs = {
    "flipper_x_body_mass": ("flipper_length_mm", "body_mass_g"),
    "bill_length_x_body_mass": ("bill_length_mm", "body_mass_g"),
    "bill_length_x_flipper": ("bill_length_mm", "flipper_length_mm"),
}
for new_col, (a, b) in interaction_pairs.items():
    df[new_col] = df[a] * df[b]

base_features = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
inter_cols = list(interaction_pairs.keys())

models = {"baseline_all4": base_features}
for r in range(1, len(inter_cols) + 1):
    for combo in combinations(inter_cols, r):
        if r == len(inter_cols):
            name = "all4 + all_3_interactions"
        else:
            name = "all4 + " + " + ".join(combo)
        models[name] = base_features + list(combo)


Z_map = {}
for name, feats in models.items():
    X = StandardScaler().fit_transform(df[feats])
    Z = linkage(X, method="complete", metric="euclidean")
    Z_map[name] = Z

    plt.figure(figsize=(10, 4))
    dendrogram(Z, no_labels=True)
    plt.title(f"Dendrogram: {name}")
    plt.xlabel("Observations")
    plt.ylabel("Distance")
    plt.show()

k_choice = {
    "baseline_all4": 9,
    "all4 + flipper_x_body_mass": 9,
    "all4 + bill_length_x_body_mass": 9,
    "all4 + bill_length_x_flipper": 10,
    "all4 + flipper_x_body_mass + bill_length_x_body_mass": 10,
    "all4 + flipper_x_body_mass + bill_length_x_flipper": 9,
    "all4 + bill_length_x_body_mass + bill_length_x_flipper": 9,
    "all4 + all_3_interactions": 9,
}

rows = []
h_tables = {}

for name in models:
    k = k_choice[name] 
    labels = fcluster(Z_map[name], t=k, criterion="maxclust")
    h = pd.crosstab(df["species"], labels)
    h_tables[name] = h

    sep, pur = score_table(h)  
    rows.append({
        "model": name,
        "k": k,
        "separation_score": sep,
        "purity": pur
    })

results = pd.DataFrame(rows).sort_values("separation_score", ascending=False).reset_index(drop=True)
display(results.round(4))

```

# Prediction

The earlier variable separation analysis addressed the question of which variables or combinations of variables best separated the species. Prediction, however, serves a slightly different purpose. When predicting species, it is important to keep not only the variables that provide the strongest overall separation, but also variables that still carry useful species information, even if they do not separate all three species equally well. This is because the separation analysis measures the overall structure of the data, while a predictive model uses variables as clues for making the best possible classification. In other words, a variable does not have to be the strongest overall separator to still improve prediction. Therefore, my final predictive model includes bill length, bill depth, flipper length, body mass, flipper length × body mass, bill length × body mass, bill length × flipper length, and island.

```python


df["flipper_x_body_mass"] = df["flipper_length_mm"] * df["body_mass_g"]
df["bill_length_x_body_mass"] = df["bill_length_mm"] * df["body_mass_g"]
df["bill_length_x_flipper"] = df["bill_length_mm"] * df["flipper_length_mm"]


num_feats = [
    "bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g",
    "flipper_x_body_mass", "bill_length_x_body_mass", "bill_length_x_flipper"
]


X_num = df[num_feats].copy()
X_island = pd.get_dummies(df["island"], prefix="island", drop_first=False)
X = pd.concat([X_num, X_island], axis=1)

y = df["species"]

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1)

acc_scores = []
f1_scores = []
bal_scores = []

for fold, (train_idx, test_idx) in enumerate(cv.split(X, y), 1):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    rf = RandomForestClassifier(
        n_estimators=500,
        random_state=1,
        class_weight="balanced"
    )
    rf.fit(X_train, y_train)

    y_pred = rf.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    f1m = f1_score(y_test, y_pred, average="macro")
    bal = balanced_accuracy_score(y_test, y_pred)

    acc_scores.append(acc)
    f1_scores.append(f1m)
    bal_scores.append(bal)

    print(f"Fold {fold}: acc={acc:.4f}, f1_macro={f1m:.4f}, bal_acc={bal:.4f}")
    
print("\nCV summary:")
print(f"Accuracy:      {np.mean(acc_scores):.4f} ± {np.std(acc_scores):.4f}")
print(f"F1 macro:      {np.mean(f1_scores):.4f} ± {np.std(f1_scores):.4f}")
print(f"Balanced Acc.: {np.mean(bal_scores):.4f} ± {np.std(bal_scores):.4f}")
```

The results obtained from the prediction model were exceptionally good for 5-fold cross-validation. The average accuracy achieved by the model was found to be 0.9880, whereas the average macro F1 was recorded at 0.9852 and the average balanced accuracy was observed at 0.9828. From these figures, it can be concluded that the model is able to predict the penguin species with very high accuracy, and this is true for all the three species of penguins.
